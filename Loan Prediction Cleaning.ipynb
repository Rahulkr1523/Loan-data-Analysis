```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\rahul\OneDrive\Desktop\PYTHON\Loan Prediction Dataset.csv")
df.shape
```

```python
# Changing the Datatype of columns as per required
df['LoanAmount'] = pd.to_numeric(df['LoanAmount'], errors='coerce')
df['Loan_Amount_Term'] = pd.to_numeric(df['Loan_Amount_Term'], errors='coerce')
df['Dependents'] = df['Dependents'].replace('3+', 3)
df['Dependents'] = pd.to_numeric(df['Dependents'], errors='coerce')
```

```python
df['Self_Employed'].fillna('No', inplace=True)              # Filling missing values in Self_Employed column
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)     # filling missing values with mode
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)

df['Gender'].fillna(df['Gender'].mode()[0], inplace=True) # Gender is Categorical so filling with mode value
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].astype(int)
df['Dependents'] = df['Dependents'].astype(int)
```
